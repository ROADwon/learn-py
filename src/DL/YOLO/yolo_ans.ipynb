```python
from ultralytics import YOLO
```

```python
model = YOLO("yolov8s.pt")
```

```python
result = model.predict("C:/Users/line/Desktop/brand_new_learn_py/data/newyork.jpg")
```

```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
plots = result[0].plot()
plots = cv.cvtColor(plots, cv.COLOR_BGR2RGB)
plt.figure(figsize=(10,10))
plt.imshow(plots)
```
